#### Varying Temperature

```python
import warnings
warnings.filterwarnings("ignore")

import gpt_2_simple as gpt2
from datetime import datetime

import os

sess = gpt2.start_tf_sess()
gpt2.download_gpt2()

gpt2.finetune(sess,
             dataset = os.path.join('data', 'transcripts.txt'),
             model_name='124M',
             steps = 1000)
```

```python
gpt2.generate(sess,
             temperature = 0.2)
```

```python
gpt2.generate(sess,
             temperature = 0.5)
```

Varying Temperature + top_k

```python
gpt2.generate(sess,
             temperature = 0.7,
             top_k = 40)
```

```python
gpt2.generate(sess,
             temperature = 0.7,
             top_k = 40)
```
